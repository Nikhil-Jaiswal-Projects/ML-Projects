```python
import pandas as pd
```

```python
data = pd.read_csv(r"ML Naïve Bayes Resource16936477090.txt")
```

```python
data.head(10)
```

```python
data.drop("Time",axis=1,inplace=True)
```

```python
pd.set_option("display.max_columns",None)
```

```python
x = data.drop("Class",axis=1)
```

```python
y = data["Class"]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale1 = StandardScaler()
```

```python
data["Amount"] = scale1.fit_transform(data["Amount"].values.reshape(-1,1))
```

```python
data.head(2)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.7,random_state=42)
```

# Navy bayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model1 = GaussianNB()
```

```python
model1.fit(x_train,y_train)
```

```python
y_pri = model1.predict(x_test)
```

```python
model1.score(x_train,y_train)*100
```

```python
model1.score(x_test,y_test)*100
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_pri,y_test))
```

#  Random forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model2 = RandomForestClassifier()
```

```python
model2.fit(x_train,y_train)
```

```python
y_pri_r = model2.predict(x_test)
```

```python
model2.score(x_train,y_train)*100
```

```python
model2.score(x_test,y_test)*100
```

```python
print(classification_report(y_pri_r,y_test))
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy_score(y_pri_r,y_test)*100
```

```python
c = confusion_matrix(y_pri_r,y_test)
```

```python
import seaborn as sns
```

```python
sns.heatmap(c,annot=True)
```
